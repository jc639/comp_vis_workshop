## 1.2 Creating Images

Lets first import all the previously imported libraries.

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
from google.colab.patches import cv2_imshow
````

The below code first creates a blank array filled with zeros with the shape (750, 750, 3) which is similar to what an image would look like in terms of shape. After that the following 3 lines will individually replace all elements of each channel with a random integer between 0 and 255. We then use the imshow method of Matplot to display the image. Again, note that this is the RGB version and what you see may be diffierent to what OpenCV would show.

````python
image = np.zeros((750, 750, 3), dtype=np.uint8)
image[:,:,0] = np.random.randint(0, 255)
image[:,:,1] = np.random.randint(0, 255)
image[:,:,2] = np.random.randint(0, 255)

plt.imshow(image)
````

Run the above command multiple times to see how the colour changes each time. This is one of the simplest method to creating blank images in different colours.

Lets now carry this code over to a new cell where we will try to change the colour of the image periodically creating a beautiful effect.

```python
import random

for i in range(100):
    image = np.zeros((750, 750, 3), dtype=np.uint8)
    image[:,:,0] = np.random.randint(0, 255)
    image[:,:,1] = np.random.randint(0, 255)
    image[:,:,2] = np.random.randint(0, 255)

    cv2_imshow('output_window', image)
    time.sleep(0.2)
    clear_output(wait=True)
```

Moving on, lets now create a circle within the blank canvas. And for this, we're going to choose specific colours for the canvas and the circle.

````python
image = np.zeros((750, 750, 3), dtype=np.uint8)
image[:,:,0] = 20
image[:,:,1] = 140
image[:,:,2] = 200
cv2.circle(image, (200,200), 20, (255,255,255), -1)

plt.imshow(image)
````

Now, lets port this code into a new cell and make the circle move from the top of the image to the bottom. Ooooo, that sounds interesting. Remember though, if you want the same blue background, you'll have to switch the values of the first and last colour channel as shown the codeblock below.

```python
for i in range(750):
    image = np.zeros((750, 750, 3), dtype=np.uint8)
    image[:,:,0] = 200
    image[:,:,1] = 140
    image[:,:,2] = 20
    cv2.circle(image, (200,i), 20, (255,255,255), -1)

    cv2_imshow('output_window', image)
    time.sleep(0.05)
    clear_output(wait=True)
```

### Big Question : Did the Ball move ?

Actually, the ball did not move. What happened instead was that at every iteration, we created a new blank canvas and drew a new circle slightly below the previous one. And when streamed sequentially, we experienced an illusion of the moving effect of the ball. This intutition is an important one if we want to understand how to work with moving images which are commonly known as videos. All video sequences that you see whether in live TV, Netflix, youtube, camcorders, casette tapes etc apply this concept to create an illusion in such a way that we experience actors and objects move within the frame as we watch them on screen. Thus, when analysing videos using computer vision, we will be analysing a sequence of images and the illusion that they create.


Moving on from the ball illusion, lets use some more OpenCV function for drawing other different shapes. As we draw these shapes, we will use the same image file that we create at the beginning and use the drawing function and the displaying multiple times. As we do that, please notice how the image object behaves. It is important to note that the draw function makes permanent changes to image that you are drawing on. So, in case, you have already drawn on an image and would want to re-draw stuff, you'd have to create the image again or reload if loading from a file path.

````python
image = np.zeros((750, 750, 3), dtype=np.uint8)
image[:,:,0] = 150
image[:,:,1] = 50
image[:,:,2] = 100
cv2.circle(image, (200,200), 20, (255,255,255), 2)
cv2.circle(image, (500,200), 20, (255,255,255), 2)
plt.imshow(image)
````

````python
cv2.rectangle(image, (300,300), (400,400), (255, 255, 255), 3)
plt.imshow(image)
````

````python
cv2.line(image, (200, 500), (500, 500), (255, 255, 255), 5)
plt.imshow(image)
````

````python
cv2.circle(image, (350,350), 300, (255,255,255), 2)
plt.imshow(image)
````

````python
# text
text = 'Happy Days Learning Computer Vision!'
  
# font
font = cv2.FONT_HERSHEY_SIMPLEX
  
# origin co-ordinates
origin = (100, 700)
  
# fontScale
fontScale = 1
   
# Red color in BGR
color = (255, 255, 255)
  
# Line thickness of 2 px
thickness = 2
   
# Using cv2.putText() method
image = cv2.putText(image, text, origin, font, fontScale, 
                 color, thickness, cv2.LINE_AA, False)

plt.imshow(image)
````
